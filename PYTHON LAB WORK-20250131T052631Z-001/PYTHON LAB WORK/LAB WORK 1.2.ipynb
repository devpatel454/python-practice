```python
# write a python program

integer_var = 10
float_var =20.5
string_var = "hello dev"
list_var = [1, 2, 3, 4, 5]
tuple_var = (10, 20, 30)
dictionary_var = {"name": "dev", "age":20}


print("integer value:", integer_var)
print("type of integer:", type(integer_var))

print("float value:", float_var)
print("type of float:", type(float_var))

print("string value:", string_var)
print("type of string:", type(string_var))

print("list value:", list_var)
print("type of list", type(list_var))

print("tuple value:", tuple_var)
print("type of tuple", type(tuple_var))

print("dictionary value:", dictionary_var)
print("type of dictionary", type(dictionary_var))
```
